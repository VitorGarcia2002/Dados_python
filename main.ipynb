## GRAFICO DOS VISITANTES QUE MAIS SOFRERAM GOLS DO CORINTHIANS NA ARENA ##

```python
import pandas as pd 
import matplotlib.pyplot as plt

coringao = pd.read_csv('./Data/Jogos.csv')

soma_mais_sofreram = coringao.groupby('VISITANTE')['GOL COR'].sum().reset_index()

soma_mais_sofreram = soma_mais_sofreram.sort_values(by='GOL COR', ascending=False).head(8)


plt.bar(soma_mais_sofreram['VISITANTE'], soma_mais_sofreram['GOL COR'],width = 0.4, color='blue', edgecolor='black')
plt.title("Visitantes que mais sofreram gols do Corinthians")

plt.ylabel("Gols Sofridos")


for i, valor in enumerate(soma_mais_sofreram['GOL COR']):
    plt.text(i, valor, str(valor), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.figure(figsize=(8, 6))
plt.tight_layout()
plt.show()
```

## GRAFICO DOS VISITANTES QUE MAIS SOFRERAM GOLS DO CORINTHIANS NA ARENA ##

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

coringao = pd.read_csv('./Data/Jogos.csv')

mais_marcaram = coringao.groupby('VISITANTE')['GOL VIS'].sum().reset_index()

mais_marcaram = mais_marcaram.sort_values(by='GOL VIS', ascending=False).head(8)


plt.bar(mais_marcaram['VISITANTE'], mais_marcaram['GOL VIS'],width = 0.4, color='red', edgecolor='black')
plt.title("Visitantes que mais marcaram")

plt.ylabel('Gols Marcados')

for i, valor in enumerate(mais_marcaram['GOL VIS']):
    plt.text(i, valor, str(valor), ha='center', va='bottom')
plt.xticks(rotation=90)
plt.figure(figsize=(8, 6))
plt.tight_layout()
plt.show()

```

## MDIA DE GOLS DOS 7 MAIORES ARTILHEIROS E GRAFICO COM A QUANTIDADE DE GOLS DOS ARTILHEIROS ##

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt


jogadores = pd.read_csv('./Data/E - Jogadores.csv')

jogadores['Mdia de Gols'] = jogadores['GOLS'] / jogadores['JOGOS']
media_gols = jogadores.sort_values(by='GOLS', ascending=False).reset_index().head(7)


plt.figure(figsize=(10, 6))
plt.scatter(media_gols['JOGOS'], media_gols['GOLS'], marker='o', c='blue', edgecolor='black',)


for i, row in media_gols.iterrows():
    plt.annotate(row['JOGADOR'], (row['JOGOS'], row['GOLS']), fontsize=12, ha='left')

plt.xlabel('Quantidade de JOGOS')
plt.ylabel('Quantidade de GOLS')
plt.title('Mdia de Gols por Jogador (Top 7)')


plt.tight_layout()
plt.show()



artilheiros = jogadores.sort_values(by='GOLS', ascending=False).reset_index(drop=True).head(7)

artilheiros.set_index('JOGADOR')['GOLS'].plot(kind='bar', color ='blue',edgecolor='black')
plt.title("Artilheiros da Neo Quimica Arena")
plt.xticks(rotation=75)
plt.figure(figsize=(8, 6))
plt.show()
```

## TODOS OS MODOS DE GOLS FEITOS NA ARENA E AS QUANTIDADES

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

goal1 = pd.read_csv('./Data/C - Gols Marcados.csv')
goal2 = pd.read_csv('./Data/D - Gols Sofridos.csv')
contagem_estilos1 = goal1['ESTILO'].value_counts().reset_index()
contagem_estilos2 = goal2['ESTILO'].value_counts().reset_index()



contagem_estilos1.columns = ['ESTILO', 'TOTAL GOLS COR']
contagem_estilos2.columns = ['ESTILO', 'TOTAL GOLS VIS']



total_gols = contagem_estilos1.merge(contagem_estilos2, on='ESTILO', how='outer')
total_gols = total_gols.fillna(0)
total_gols['TOTAL GOLS'] = total_gols['TOTAL GOLS COR'] + total_gols['TOTAL GOLS VIS']

print(total_gols)


plt.figure(figsize=(8, 6))
plt.pie(total_gols['TOTAL GOLS'], startangle=140)
plt.title("Modo de gols")
plt.axis('equal')  

plt.legend(total_gols['ESTILO'] + ': ' + total_gols['TOTAL GOLS'].astype(int).astype(str), loc='center left', bbox_to_anchor=(1, 1))
plt.show()
```

## JOGADORES DO CORINTHIANS COM A MAIOR TAXA DE VITRIA NA ARENA ##

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

jogadores_cor = pd.read_csv('./Data/E - Jogadores.csv')
jogadores_cor = jogadores_cor[jogadores_cor['JOGOS']>50]
jogadores_cor['TAXA VITRIA']= jogadores_cor['VITRIAS']/(jogadores_cor['JOGOS'])*100

melhor_aproveitamento = jogadores_cor.sort_values(by='TAXA VITRIA', ascending = False).reset_index().head(10)


print(melhor_aproveitamento[['JOGADOR','JOGOS', 'VITRIAS','TAXA VITRIA' ]])

plt.figure(figsize=(8, 6))
plt.pie(melhor_aproveitamento['TAXA VITRIA'], startangle=140)
plt.title("Jogadores com maior Taxa de Vitria na Arena")
plt.axis('equal')  

plt.legend(melhor_aproveitamento['JOGADOR'] + ': ' + melhor_aproveitamento['TAXA VITRIA'].astype(int).astype(str) + "%", loc='center left', bbox_to_anchor=(1, 1))
plt.show()

```

## TIMES QUE MAIS PERDERAM PRO CORINTHIANS E ##
## TIMES QUE MAIS VENCERAM DO CORINTHIANS ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

Jogos = pd.read_csv("./Data/Jogos.csv")
vitorias_corinthians = Jogos[Jogos['RESULTADO'] == 'V']
derrotas_corinthians = Jogos[Jogos['RESULTADO'] == 'D']

fregues_corinthians = vitorias_corinthians.groupby('VISITANTE').size().reset_index(name='NUM DERROTAS')
fregues_corinthians = fregues_corinthians.sort_values(by='NUM DERROTAS', ascending=False).head(9)

carrasco_corinthians = derrotas_corinthians.groupby('VISITANTE').size().reset_index(name='NUM VITORIAS')
carrasco_corinthians = carrasco_corinthians.sort_values(by='NUM VITORIAS', ascending=False).head(9)


fregues_corinthians.set_index('VISITANTE')['NUM DERROTAS'].plot(kind='bar', color ='black',edgecolor='black') 
plt.title("Times que mais perderam")
plt.xticks(rotation=75)
plt.figure(figsize=(8, 6))
plt.show()

carrasco_corinthians.set_index('VISITANTE')['NUM VITORIAS'].plot(kind='bar', color ='green',edgecolor='black') 
plt.title("Times que mais Venceram")
plt.xticks(rotation=75)
plt.figure(figsize=(8, 6))
plt.show()



```
